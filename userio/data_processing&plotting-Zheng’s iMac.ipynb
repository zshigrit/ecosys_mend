```python
## 3-1 process model outputs: soil carbon: all ten years
pd.set_option('display.max_columns', None)

for inp_level in np.arange(6):
    for N_level in np.arange(6):
        df = pd.read_csv(f'./out_inp{inp_level}_n{N_level}/aCaN_VAR_year.out',skiprows=1,delim_whitespace=True)
#         df.iloc[1:,:].plot(x='Year',y='SOM_C')
        plt.plot(df.iloc[1:,:].Year,df.iloc[1:,:].SOM_C,label=f'inp: level{inp_level} AND N: level{N_level}')
#         plt.legend()
```

```python
## 3-2 process model outputs: soil carbon: last three year mean
df_soc = pd.DataFrame(columns=['INP_levels','N_levels','SOC'])
for inp_level in np.arange(6):
    for N_level in np.arange(6):
        df = pd.read_csv(f'./out_inp{inp_level}_n{N_level}/aCaN_VAR_year.out',skiprows=1,delim_whitespace=True)
        df_soc = df_soc.append({'INP_levels':inp_level,'N_levels':N_level,
                                'SOC':np.mean(df['SOM_C'].iloc[-3:])},ignore_index=True)
        
fig = plt.figure(figsize=(12,12))
for inp_level in np.arange(6):        
    ax=fig.add_subplot(3,2,inp_level+1)
    df_soc[df_soc.INP_levels==inp_level]['SOC'].plot.bar(ax=ax)
plt.tight_layout()

plt.tight_layout()
# df_soc[df_soc.N_levels==0.0]['SOC'].plot.bar()
```

```python
## 3-3 process model outputs: soil carbon: last three year mean
df_soc = pd.DataFrame(columns=['INP_levels','N_levels','SOC'])
for inp_level in np.arange(6):
    for N_level in np.arange(6):
        df = pd.read_csv(f'./out_inp{inp_level}_n{N_level}/aCaN_VAR_year.out',skiprows=1,delim_whitespace=True)
        df_soc = df_soc.append({'INP_levels':inp_level,'N_levels':N_level,
                                'SOC':np.mean(df['SOM_C'].iloc[-3:])},ignore_index=True)

fig = plt.figure(figsize=(12,12))
for n_level in np.arange(6):        
    ax=fig.add_subplot(3,2,n_level+1)
    df_soc[df_soc.N_levels==n_level]['SOC'].plot.bar(ax=ax)
plt.tight_layout()
# df_soc[df_soc.N_levels==0.0]['SOC'].plot.bar()
```

```python
## 3-4 process model outputs: soil carbon: last three year mean

# df_soc = pd.DataFrame(columns=['INP_levels','N_levels0','N_levels1','N_levels2',
#                                'N_levels3','N_levels4','N_levels5'])
df_soc = pd.DataFrame(columns=['INP_levels','Nfix_levels','SOC'])
for inp_level in np.arange(6):
    for N_level in np.arange(6):
        df = pd.read_csv(f'./out_inp{inp_level}_n{N_level}/aCaN_VAR_year.out',skiprows=1,delim_whitespace=True)
        df_soc = df_soc.append({'INP_levels':inp_level,'Nfix_levels':N_level,
                                'SOC':np.mean(df['SOM_C'].iloc[-3:])},ignore_index=True)
        
fig,ax = plt.subplots()
for key,grp in df_soc.groupby('INP_levels'):
    ax = grp.plot(ax=ax, kind='line', x='Nfix_levels',y='SOC',label=f'Input_level{int(key)}',linestyle='-', marker='o')
    
ax.legend(bbox_to_anchor=(1.1, 1.05))    
ax.set_ylabel('SOC (kg/m$^2$)')

plt.tight_layout()

plt.savefig('./figures/figure1_soc.png',dpi=150)
```

```python
## 3-10 process model outputs: soil respiration last three year mean


df_var = pd.DataFrame(columns=['INP_levels','Nfix_levels','Rs'])
for inp_level in np.arange(6):
    for N_level in np.arange(6):
        df = pd.read_csv(f'./out_inp{inp_level}_n{N_level}/aCaN_FLX_year.out',skiprows=1,delim_whitespace=True)
        df_var = df_var.append({'INP_levels':inp_level,'Nfix_levels':N_level,
                                'Rs':np.mean(df['CO2_soil'].iloc[-3:])},ignore_index=True)
        
# df_var['Rs'] = df_var.Rs*10.**3*365.*24. ## g/m3/yr from(mg/cm3/h)
df_var['Rs'] = df_var.Rs*(10**9)/(12*3600) ## umol/m2/s

fig,ax = plt.subplots()
for key,grp in df_var.groupby('INP_levels'):
    ax = grp.plot(ax=ax, kind='line', x='Nfix_levels',y='Rs',label=f'Input_level{int(key)}',linestyle='-', marker='o')
    
ax.legend(bbox_to_anchor=(1.1, 1.05))    
ax.set_ylabel(r'Rs ($\mu$mol m$^{\rm -2}$ s$^{\rm -1}$)')  # plt.xlabel(r'Primary T$_{\rm eff}$')
ax.ticklabel_format(useOffset=False, style='plain')

plt.tight_layout()

plt.savefig('./figures/figure6_Rs.png',dpi=150)
```
